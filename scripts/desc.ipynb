teste de normalidade

Um gráfico QQ é uma ferramenta gráfica que nos ajuda a avaliar se um conjunto de dados segue uma distribuição teórica específica. Neste caso, está verificando a distribuição normal. O eixo x mostra os quantis teóricos, o que significa o que os dados deveriam seguir se fossem normalmente distribuídos. O eixo y mostra os quantis dos dados reais. Se os dados forem distribuídos normalmente, os pontos no gráfico QQ devem cair aproximadamente em uma linha reta diagonal.

A função normaltest testa a hipótese nula de que uma amostra vem de uma distribuição normal. Ele retorna dois valores: k2 (a estatística de teste) e p (o valor p). Um valor p mais baixo significa que há menos probabilidade de os dados virem de uma distribuição normal.

O alfa é definido como 0,05, que é um limite comum para significância estatística. Se o valor p for menor que alfa, o script imprime que a hipótese nula pode ser rejeitada, o que significa que os dados não são distribuídos normalmente. Se o valor p for maior que alfa, isso significa que a hipótese nula não pode ser rejeitada, o que significa que os dados podem ser distribuídos normalmente.

Em resumo, este script verifica a normalidade de cada coluna de dados especificada, tanto visualmente (com o gráfico QQ) quanto estatisticamente (com o teste de normalidade).

```python
import matplotlib.pyplot as plt
from scipy.stats import probplot, normaltest

columns = ['PL', 'PC', 'ARLP', 'VLL']

for col in columns:
    plt.figure(figsize=(10, 4))

    # Create a Q-Q plot
    plt.subplot(1, 2, 1)
    probplot(df[col], plot=plt)
    plt.title(f'Q-Q plot of {col}')

    # Perform the normality test
    k2, p = normaltest(df[col])
    alpha = 0.05
    print(f'p = {p}')
    if p < alpha:  # null hypothesis: x comes from a normal distribution
        print(f'The null hypothesis can be rejected for {col}')
    else:
        print(f'The null hypothesis cannot be rejected for {col}')

    plt.show()
```

```python
from scipy.stats import kstest
from scipy.stats import norm
import pandas as pd

# Create an empty DataFrame to store the results
results = pd.DataFrame(columns=['Column', 'Statistic', 'p-value'])

for col in columns:
    # Standardize the column to have mean=0 and variance=1
    standardized_col = (df[col] - df[col].mean()) / df[col].std()

    # Perform the Kolmogorov-Smirnov test
    statistic, p_value = kstest(standardized_col, 'norm')

    # Append the results to the DataFrame
    new_row = pd.DataFrame({'Column': [col], 'Statistic': [statistic], 'p-value': [p_value]})
    results = pd.concat([results, new_row], ignore_index=True)

# Display the results
print(results)
```

O teste Kolmogorov-Smirnov retorna dois valores: a estatística de teste (D) e o valor p. A estatística de teste é a diferença máxima entre a função de distribuição cumulativa empírica da amostra e a função de distribuição cumulativa da distribuição de referência. O valor p é a probabilidade de observar uma estatística de teste tão extrema quanto a calculada, assumindo que a hipótese nula seja verdadeira.

A coluna 'Estatística' no DataFrame de resultados corresponde à estatística de teste (D) e a coluna 'valor p' corresponde à Significância Assintótica (bicaudal) ou valor p.

```python
# Rename the columns
results.columns = ['Column', 'Kolmogorov-Smirnov Z', 'Asymp. Sig (2-tailed)']

# Display the results
print(results)
```

O teste de Lilliefors é um teste estatístico usado para verificar se os dados vêm de uma população normalmente distribuída. É uma variante do teste de Kolmogorov-Smirnov e é utilizado quando os parâmetros (média e desvio padrão) da distribuição normal não são conhecidos, mas são estimados a partir dos dados.

O teste funciona comparando a função de distribuição empírica dos dados amostrais com a função de distribuição cumulativa da distribuição normal. A hipótese nula (H0) do teste é que os dados provêm de uma distribuição normal. Se o valor p obtido no teste for menor que o nível de significância escolhido (geralmente 0,05), então a hipótese nula é rejeitada e conclui-se que os dados não provêm de uma distribuição normal.

```python
from statsmodels.stats.diagnostic import lilliefors
import pandas as pd

# Create an empty DataFrame to store the results
results = pd.DataFrame(columns=['Column', 'Statistic', 'p-value'])

for col in columns:
    # Perform the Lilliefors test
    statistic, p_value = lilliefors(df[col])

    # Append the results to the DataFrame
    new_row = pd.DataFrame({'Column': [col], 'Statistic': [statistic], 'p-value': [p_value]})
    results = pd.concat([results, new_row], ignore_index=True)

    # Print the result of the hypothesis test
    if p_value > 0.05:
        print(f'Para {col}, o p-valor é maior que 0.05. Não rejeite H0, portanto a distribuição é normal.')
    else:
        print(f'Para {col}, o p-valor é menor que 0.05. Rejeite H0, portanto a distribuição não é normal.')

# Display the results
results
```

### Transformações para obter normalidade

• Distribuição assimétrica positiva: emprega-se o logaritmo das variá-

veis. Distribuição assimétrica negativa: emprega-se a raiz quadrada das va- riáveis.

• Distribuição achatada: emprega-se o inverso das variáveis (1/y e 1/x).

Transformações para obter homoscedasticidade

• Distribuição dos resíduos: empregam-se logaritmo, raiz quadrada, in- verso etc.

• A transformação deverá ser testada para verificar se o remédio utiliza- do é eficiente.

Transformações para obter linearidade

• De acordo com Matos (1997, p. 32), existem as seguintes formas para obter linearidade:


### Tipo de função

* Logarítmica ou potencial 

original:
Y=a.X^{b}

Forma linearizada por transformação
LnY = Lna + b.LnX

* Exponencial ou semilogarítmica I

original:
Y=a.b^{x}

Forma linearizada por transformação
LnY = Lna + Lnb.X

* Semilogarítmica II

original:
e^{y}=a.X^{b}

Forma linearizada por transformação
Y=Lna+b.LnX

* Hiperbólica ou recíproca I

original:
Y=a+b1/X

Forma linearizada por transformação
Usa-se 1/X em vez de X


* Hiperbólica ou recíproca II

original:
Y=1/(a+bX)

Forma linearizada por transformação
1/Y=a+bX

* Quadrática

original:
Y=a+bX+cX^{2}

Forma linearizada por transformação
Usa-se X² além de X

* Logística

original:
Y=M+b.e^{-mx}

Forma linearizada por transformação
Ln(M/Y-1)=Lnb-mX
